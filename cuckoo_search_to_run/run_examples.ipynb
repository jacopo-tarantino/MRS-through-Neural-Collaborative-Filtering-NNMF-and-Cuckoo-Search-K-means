## Functions

### Himmelblau  
$f(x,y) = (x^2+y-11)^2+(x+y^2-7)^2$

```python
# setting the fucntion
himmelbau=Function()
himmelbau.set_expression(lambda z:(z[0,0]**2 + z[0,1] - 11)**2 + (z[0,0] + z[0,1]**2 - 7)**2)
himmelbau.set_function_params(shape=[1,2],BoundUpper=[6,6],BoundLower=[-6,-6])
```

```python
# running the algorithm
cs=cuckoo_search(cuckoosNumber=5,hostNumber=5,function=himmelbau,iterNumber=1000,lmbd=1.8,alpha=0.001,stored_iter=10)
cs.run()
# plotting
cs.counture_plot(levels=100)
```

### Ackley  
$f(x,y)=-20exp(-0.2\sqrt{0.5(x^2+y^2)})-exp(0.5(cos(2\pi x)+cos(2\pi y)))+e+20$

```python
# setting the function
ackley = Function()
ackley.set_expression(lambda z: -20*math.exp(-0.2*math.sqrt(0.5*(z[0,0]**2+z[0,1]**2))) - math.exp(0.5*(math.cos(2*math.pi*z[0,0])+math.cos(2*math.pi*z[0,1])))+ math.e + 20)
ackley.set_function_params(shape=[1,2], BoundUpper=[4,4],BoundLower=[-4,-4])
```

```python
# running the algorithm
cs=cuckoo_search(cuckoosNumber=5,hostNumber=5,function=ackley,iterNumber=1000,lmbd=1.8,alpha=0.001,stored_iter=10)
cs.run()
# plotting
cs.counture_plot(levels=50)
```
